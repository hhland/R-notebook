```r
iris<-read.csv("http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",header = F)

colnames(iris)<-c("sepal_length","sepal_width","petal_length","petal_width","cls")


head(iris)
```

```r
myPCA<- function(df){
    
    eig<- eigen(cov(df))
    means<-unlist(lapply(df,mean))
    scores<- scale(df,center = means) %*% eig$vectors
    list(values=eig$values,vectors=eig$vectors,scores=scores)
}
```

```r
my_pca<- myPCA(iris[1:4])

head(my_pca)
```

```r
pca <- princomp(iris[1:4],scores = T)
cbind(unlist(lapply(my_pca[1],sqrt)),pca$sdev)
```

```r
summary(pca)
```
