   </tr>

      <td>表示变量的交互项。例如，用<code>A</code>、<code>B</code>和<code>A</code>与<code>B</code>的交互项来预测<code>y</code>，代码为<code>y~ A + B + A:B</code></td>
   </tr>
   <tr>
      <td align="center"><code>*</code></td>
      <td>表示所有可能交互项。代码<code>y~ A * B * C</code>可展开为<code>y ~ A + B + C + A:B + A:C + B:C + A:B:C</code></td>
   </tr>
   <tr>
      <td align="center"><code>^</code></td>
      <td>表示交互项达到某个次数。代码<code>y ~ (A + B + C)^2</code>可展开为<code>y ~ A + B + C + A:B + A:C + B:C</code></td>
   </tr>
   <tr>
      <td align="center"><code>.</code></td>
      <td>表示包含除因变量外的所有变量。例如，若一个数据框包含变量<code>y</code>、<code>A</code>、<code>B</code>和<code>C</code>，代码<code>y ~ .</code>可展开为<code>y ~ A + B + C</code></td>
   </tr>

#One-way ANOVA

```python
%%R

library(multcomp)
attach(cholesterol)

table(trt)
```

```python
%R aggregate(response, by=list(trt), FUN=mean) 
```

```python
%R  aggregate(response, by=list(trt), FUN=sd) 
```

```python
%%R 

fit <- aov(response ~ trt)

summary(fit)
```

```python
%%R

library(gplots)

plotmeans(response ~ trt, xlab="Treatment", ylab="Response",main="Mean Plot\nwith 95% CI")

detach(cholesterol)
```

```python
%%R

#detach("package::HH")
TukeyHSD(fit)
```

```python
%%R

par(las=2)
par(mar=c(5,8,4,2))
plot(TukeyHSD(fit))
```

```python
%%R
library(multcomp)

par(mar=c(5,4,6,2))
tuk <- glht(fit, linfct=mcp(trt="Tukey"))
plot(cld(tuk, level=.05),col="lightgrey")
```

```python
%%R
library(car)
qqPlot(lm(response ~ trt, data=cholesterol),simulate=TRUE, main="Q-Q Plot", labels=FALSE)
```

```python
%R bartlett.test(response ~ trt, data=cholesterol)
```

```python
%%R

library(car)
outlierTest(fit)
```
